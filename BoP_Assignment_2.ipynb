```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

print("Given three lines are : a1x + b1y + c1 = 0 ; a2x + b2y + c2 = 0 ; a3x + b3y + c3 = 0")
print("Let us consider three lines as :\n3x - 2y - 1 = 0")
print(" x - 4y + 3 = 0")
print(" x +  y - 7 = 0")
print("Therefore we can say three points that form a triangle are : (x1,y1)=(1,1) , (x2,y2)=(3,4) , (x3,y3)=(5,2)")

#therefore for above points we have formula for area of triangle as 0.5*abs[x1y2-x2y1+x2y3-x3y2+x3y1-y1x3]

print("Using formula area of triangle as : 0.5*abs([x1y2-x2y1+x2y3-x3y2+x3y1-y1x3]) we get ,")
x1=1
x2=3
x3=5
y1=1
y2=4
y3=2

val=(x1*y2 - x2*y1 + x2*y3 - x3*y2 + x3*y1 - x1*y3)
area_of_triangle=0.5*abs(val)
print("\nArea of triangle as :",area_of_triangle,"\n")

print("Now comparing above equations with aix + biy + ci = 0 , we get")
print("a1=3,b1=-2,c1=-1\na2=1,b2=-4,c2=3\na3=1,b3=1,c3=-7")

print("Now to find determinant of \n|a1,b1,c1|")
print("|a2,b2,c2|")
print("|a3,b3,c3|")
#finding determinant of matrix of above equations
a = np.array([[3,-2,-1],[1,1,-7],[1,-4,3]]) 
print("\nDeterminant of above 3x3 matrix :",abs(int(np.linalg.det(a))))
xval=abs(int(np.linalg.det(a)))**2
print("Therefore determinant^2 is :",xval)
cofactor1=(1-(-4))
cofactor2=-(3-(-2))
cofactor3=(-12-(-2))
print("Now cofactors of c1,c2,c3 are:",cofactor1,cofactor2,cofactor3)
print("Therefore, (determinant^2)/(2*cofactor1*cofactor2*cofactor3) is : ",xval/(2*cofactor1*cofactor2*cofactor3))
print("Above result is equal to area of triangle , hence proved")

x = np.linspace(0, 10, 100)
y = (3/2)*x - (1/2)
fig = plt.figure(figsize = (10, 5))
plt.plot(x, y)
x = np.linspace(0, 10, 100)
y = (1/4)*x + (3/4)
plt.plot(x, y)
x = np.linspace(0, 10, 100)
y = (-1)*x + (7)
plt.plot(x, y)
plt.show()
```
